```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from pandas.core.common import random_state
from sklearn.linear_model import LinearRegression
```

```python
df_gold=pd.read_csv("kapeel dataset.csv")
df_gold.head()
```

```python
df_gold.describe()
```

```python
plt.title('YEAR')
sns.histplot(df_gold['Gold Price'])
plt.show()
```

```python
plt.scatter(df_gold['YEAR'], df_gold['Gold Price'], color = 'r')
plt.title('year vs gold')
plt.xlabel('YEAR')
plt.ylabel('Gold Price')
#plt.box(False)
plt.show()
```

```python
X = df_gold[['YEAR']]  
y = df_gold['Gold Price'] 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
y_pred_train = regressor.predict(X_train)
plt.scatter(X_train, y_train, color = 'r')
plt.plot(X_train, y_pred_train, color = 'b')
plt.title('YEAR vs Gold Price (Training Set)')
plt.xlabel('YEAR')
plt.ylabel('Gold Price')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'Gold Price/YEAR', loc='best', facecolor='green')
plt.box(False)
plt.show()
```

```python
y_pred_train = regressor.predict(X_train)
plt.scatter(X_test, y_test, color = 'r')
plt.plot(X_train, y_pred_train, color = 'b')
plt.title('YEAR vs Gold Price (Test Set)')
plt.xlabel('YEAR')
plt.ylabel('Gold Price')
plt.legend(['X_train/Pred(y_test)', 'X_train/y_train'], title = 'YEAR/Gold', loc='best', facecolor='green')
plt.box(False)
plt.show()
```

```python
print(f'Coefficient: {regressor.coef_}')
print(f'Intercept: {regressor.intercept_}')
```
